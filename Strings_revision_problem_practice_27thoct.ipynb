```python
data = {"name": "Alice", "age": 30, "city": "New York"}
template = "Name: {name}, Age: {age}, City: {city}."
formatted_string = template.format_map(data)
print(formatted_string)
```

```python
# Define a dictionary with data
user_data = {
    'name': 'Alice',
    'age': 30,
    'city': 'New York'
}

# Define a format string with placeholders
format_string = "Hello, my name is {name}, I am {age} years old, and I live in {city}."

# Use format_map() to substitute the placeholders with values from the dictionary
formatted_output = format_string.format_map(user_data)

# Print the result
print(formatted_output)
```

# A customer support bot receives messy inputs. You want to clean up the text and check if
# it contains keywords like "refund".

```python
customer_input= input()
customer_input_clean=customer_input.strip().casefold()
print(customer_input_clean)
if "refund" in customer_input_clean:
    print("Refund applicable")
```

```python
#vowel count
#char to ascii
#ascii to char
```

```python
vowels='aeiou'
count=0
string_input=input()
for i in string_input.strip().casefold():
    if i in vowels: #membership operator
        count+=1
print("Number of vowels in the string",count)
```

# 1.decorator to log time

```python
# decorator explanation
import time
def time_logger(func):
    def wrapper(*args,**kwargs):
        #log the start time
        print("LOG: arguments",args)
        print("LOG: keyword arguments",kwargs)
        start_time=time.time()
        result=func(*args,**kwargs)
        end_time=time.time()
        total_time= end_time-start_time
        print(f"LOG: total time taken, {total_time:.2f}")
        return result
    return wrapper
        
        
```

```python
@time_logger
def greet_guest(name,greeting="Hello"):
    print(f"{greeting},{name}!")
    time.sleep(2) # suspend the execution for 2 seconds
```

```python
greet_guest("Bhushan","Namaste")
```

# 2.detecting and generating primes

```python
import math
def is_prime(n):
    if n==2:
        return True
    if n<2 or n%2==0:
        return False
    else:
        #print("the input is ",n)
        #print("limit is ",int(math.sqrt(n)))
        for i in range(3,int(math.sqrt(n))+1,2):

            if n%i==0:
                return False
    return True
        
```

```python
is_prime(33)
```

```python
# generate the primes
number_terms=int(input("Number of primes?"))
```

```python
def generate_primes(n):
    """to generate primes upto n terms"""
    current_prime=2
    current_count=1
    while current_count<=n:
        if is_prime(current_prime):
            print(current_prime, end=" ")
            current_count+=1
        current_prime+=1
```

```python
generate_primes(7)
```

# 3. case study
# customer feedback analyzer
# You work with a retail company to analyze customer feedback.
# Each feedback has a name, rating (1â€“5), and comment.

```python
feedbacks = [
{"name": "Alice", "rating": 5, "comment": "Great service"},
{"name": "Bob", "rating": 3, "comment": "Averageexperience"},
{"name": "Chetan", "rating": 2, "comment": "Too slow"},
{"name": "Divya", "rating": 4, "comment": "Nice store"},
{"name": "Eshan", "rating": 1, "comment": "Bad service"}
]
```

```python
def classify_feedback(rating):
    """this processes the rating and outputs the sentiment"""
    if rating>=4:
        return "Positive"
    elif rating==3:
        return "Neutral"
    else:
        return "Negative"
```

```python
# output is in the form of a dictionary [ list of dictionaries : [{"Alice":"Positive"},{"Bob":"neutral"}... ]
#map transform in the form of sentiment

list_sentiments=list(map(lambda f:classify_feedback(f["rating"]),feedbacks))
```

```python
list_names=list(map(lambda f:f["name"],feedbacks))
list_names
```

```python
sentiments_dict=dict(zip(list_names,list_sentiments))
```

# 4.leap year checker

```python
# input is year , output is True or False
# print it is a leap year , reason
def leap_year_checker(year):
    """checks if a year is leap year or not"""
    if year%400==0: #for a century to be a leap year it must be divisible by 400
        print(f"Yes {year} is a leap year")
        print("Reason: It is divisible by 400")
        return True
    elif year%100==0 and year%400!=0:
        print(f"No {year} is not a leap year")
        print("Reason: It is divisible by 100 but not by 400")
        return False
    elif year%4==0:
        print(f"Yes {year} is a leap year")
        print("Reason: It is divisible by 4")
        return True    
    else:
        print(f"No {year} is not a leap year")
        return False
        
```

```python
list_years=[1900,2024,2020,2000,2025]
for i in list_years:
    print(leap_year_checker(i))
```

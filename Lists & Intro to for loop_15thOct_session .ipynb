### Understanding `range()` in Python
The `range()` function generates a sequence of numbers.
- Used commonly in loops.
- Syntax: `range(start, stop, step)`
- If only one argument is given, it's treated as the stop.

```python



````python
l1=range(0,10,2)
for i in l1:
    print(i)
````

````python
l2=[1,2,3,4,5]
for i in l2:
    print(i)
````

Useful for controlled iteration, repeating tasks, or indexing.

### Understanding `enumerate()` in Python
`enumerate()` helps access both index and value during iteration.
- Syntax: `enumerate(iterable, start=0)`
- It returns pairs of (index, item)




````python
for i,val in enumerate(range(0,10,2)):
    print('index',i)
    print('value',val)
````

**Use Case:** Tracking index when looping over elements.

# 0 - 2.5 - 0%
# 2.5 - 5 - 5%
# 5. - 10 - 20%
# 10 or above - 30%

```python
taxable_income = float(input('Enter taxable income'))
```

```python
taxable_income_lk = taxable_income/100000
print("taxable income in lakhs",taxable_income_lk)
#rules if elif else
if taxable_income_lk >= 10:
    print("Tax slab is {} and tax payable is {:.2f}".format('30%',taxable_income_lk*0.3))
elif taxable_income_lk >= 5 and taxable_income_lk < 10:
    print("Tax slab is {} and tax payable is {}".format('20%',taxable_income_lk*0.2))
elif taxable_income_lk >= 2.5 and taxable_income_lk < 5:
    print("Tax slab is {} and tax payable is {}".format('5%',taxable_income_lk*0.05))
else:
    print("Tax slab is {} and tax payable is {}".format('0%',taxable_income_lk*0))
          
```

```python
#match case example
user_input = input("What is your favorite place")
match user_input.strip().casefold():
    case "pune":
        print("Favorite place is Pune")
    case "delhi":
        print("Favorite place is Delhi")
    case _ : # catchall if none of the above cases are satisfied
        print("Favorite place is home")
        
```

# while else example

# Only when entire while loop executes , the else statement will get executed

```python
# suppose we are printing numbers from 0 to 10 and it breaks if the iterator value is =5
count=0
while count<=10:
    print(count)
    if count==5:
        pass
    count+=1
else:
    print("whole while loop has executed")
```

# print character count in a word using dictionary comprehension

```python
#"Innomatics", {i:2, n:2,o:1,m:1,a:1,t:1,i:1,c:1,s:1}
user_input=input("Enter the word to get the count").strip().lower()
```

```python
{ch:user_input.count(ch) for ch in user_input}
```

# fibonacci series

```python
# using list inside for 
list_fib=[0,1]
n=int(input("How many terms"))
#count=0
#for loop will run n-2 times
for i in range(1,n-1): # this runs n-2 times
    sum_prev=list_fib[i]+list_fib[i-1]
    list_fib.append(sum_prev)
print(list_fib)
```

```python
# without list , swapping the values
a=0
b=1
n=int(input("How many terms"))
for i in range(n):
    print(a,end=" ")
    a,b=b,a+b
    
```

```python
# while
n=int(input("How many terms"))
a=0
b=1
count=1
while count<=n:
    print(a,end=" ")
    a,b=b,a+b
    count+=1
```

# random guessing game

```python
import random
random_number_gen=random.randint(0,10) #upper bound is considered as well, 0-10
random_number_gen
```

```python
# 3 trials and we match the user input with the system generated number
```

```python
flag=False
for i in range(3):
    user_input=int(input("Enter your guess"))
    if user_input==random_number_gen:
        print("Guessed it right")
        flag=True
        break

if not flag:
        print("All guesses in the trials are wrong")
        
              
```

```python
# try this with while
count=0
while count<3:
    user_input=int(input("Enter your guess"))
    if user_input==random_number_gen:
        print("Guessed it right")
        flag=True
        break
    count+=1 #1,
if not flag:
        print("All guesses in the trials are wrong")
```

5. Write a program that checks if a number is divisible by 3 and 5. Print "FizzBuzz" if divisible by both, "Fizz" if only by 3, "Buzz" if only by 5, otherwise print the number itself.


```python
data = int(input("Enter the number :"))
if data%3==0 and data%5==0:
    print("FizzBuzz")
elif data%3==0:
    print("Fizz")
elif data%5==0:
    print("Buzz")
else:
    print(data)
    
```

8. Input a number and print whether it is "Positive Even", "Positive Odd", "Negative Even", "Negative Odd", or "Zero" using nested conditionals.


```python
num = int(input("Enter a number: "))

if num == 0:
    print("Zero")
else:
    if num > 0:
        if num % 2 == 0:
            print("Positive Even")
        else:
            print("Positive Odd")
    else:
        if num % 2 == 0:
            print("Negative Even")
        else:
            print("Negative Odd")
```

10. input the color of a traffic light and print "Stop" for red, "Slow down" for yellow, "Go" for green, or "Invalid color" for any other input.

```python
Traffic_light= input("Enter the traffic light color").casefold()
if Traffic_light == "red":
    print("Stop")
elif Traffic_light=="yellow":
    print("Slow down")
elif Traffic_light=="green":
    print("GO")
else:
    print("Invalid Color")
```

### Q.6 Write a function factorial(n) that returns the factorial of a given number using recursion.



```python
def factorial(n):
    if n == 0 or n == 1:
        return 1
    return n * factorial(n - 1)

 
factorial(5)
```

### Q.7 Write a function fibonacci(n) that returns the first n numbers in the Fibonacci sequence.


```python
def fib(n):
    a,b=0,1
    seq=[]
    for _ in range(n):
        seq.append(a)
        a,b=b,a+b 
    return seq 
 
fib(10)
```

### Q.9 Write a function palindrome(s) that checks whether a string is a palindrome (case-insensitive).



```python
def palindrome(s):
    s = s.replace(" ", "").lower() 
    return s == s[::-1]

data="moM"
palindrome(data)
```


### Q.11 Write a function custom_map(func, lst) that mimics the built-in map() function.



```python
def custom_map(func, lst):
    result = []
    for item in lst:
        result.append(func(item))
    return result

def square(x):
    return x * x

print(custom_map(square, [1, 2, 3, 4]))
```

### Q.12 Write a function flatten_list(nested_list) that flattens a list of lists into a single list.


```python
def flatten_list(nested_list):
    flat = []
    for sublist in nested_list:
        for item in sublist:
            flat.append(item)
    return flat

data = [[1, 2], [3, 4], [5, 6]]
print(flatten_list(data))
```


### Q.13 Write a function apply_discount(price, discount=10) that applies a discount (default 10%) and returns the new price.



```python
def apply_discount(price, discount=10):
    return price - (price * discount / 100)

print(apply_discount(50))        
print(apply_discount(200, 20))    
```

### Q.14 Write a lambda function to sort a list of tuples by the second element.

```python
sorted_list = lambda lst: sorted(lst, key=lambda x: x[1])
data = [(1, 5), (3, 1), (2, 4)]

print(sorted_list(data))
```

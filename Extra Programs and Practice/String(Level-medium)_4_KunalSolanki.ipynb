### 2. **Check for Palindrome**  
   Write a function that checks whether a given string is a palindrome (reads the same forwards and backwards).  
   *Example:* `"madam" → True"`



```python
data="madam"
reversed=data[::-1]
print(data==reversed)
```

### 3. **Count Vowels in a String**  
   Write a program that counts the number of vowels in a given string.  
   *Example:* `"OpenAI" → 3"`



```python
data="OpenAI"
vowels="aeiou"
data=data.casefold()
print(sum(data.count(v) for v in vowels))
```

### 4. **Remove Punctuation from a String**  
   Remove all punctuation from a string using Python.  
   *Example:* `"Hello, world!" → "Hello world"`  



```python
s = "Hello, world!"
punct = ",.!?:;-''"

result = ''.join(x for x in s if x not in punct)

print(result)
# for i in s:
#     if i not in punct:
#         print(i,end="")
        

```

### 6. **Find the Longest Word in a Sentence**  
   Given a string sentence, find and return the longest word.  
   *Example:* `"I love artificial intelligence" → "intelligence"`  



```python
data= "I love artificial intelligence"
data=data.split()
longest = max(data, key=len)
print(longest)
```

### 7. **Character Frequency Count**  
   Write a program that returns a dictionary of character counts in a string.  
   *Example:* `"hello" → {'h': 1, 'e': 1, 'l': 2, 'o': 1}`  


```python
data="hello"
count_data={x:data.count(x) for x in data}
print(count_data)
```

### 8. **Check If Two Strings are Anagrams**  
   Write a function to check if two strings are anagrams of each other.  
   *Example:* `"listen"` and `"silent"` → True  


```python
def is_anagram(s1, s2):
    return sorted(s1) == sorted(s2)

print(is_anagram("listen", "silent"))

# data1="listen"
# data2="silent"
# if sorted(data1)==sorted(data2):
#     print(True)
# else:
#     print(False)


```

### 10. **Find All Words Starting with a Vowel**  
    Write a program to extract all words that start with a vowel from a sentence.  
    *Example:* `"An eagle is over the ocean"` → `['An', 'eagle', 'is', 'over', 'ocean']`

```python
data ="An eagle is over the ocean"
vowels="AaEeIiOoUu"
list1=[]
data=data.split()
for i in data:
    if i[0] in vowels:
        list1.append(i)
print(list1)
```

### Q.1 Sum of N Numbers (While Loop) Write a program to calculate the sum of the first N natural numbers using a while loop.

```python
list1=[]
n=1
sum1=0
while n<11:
    sum1+=n
    list1.append(n)
    n+=1
print(f"The sum of data {list1} is",sum1)
```

### 2. Factorial of a Number (For Loop) Write a program that takes an integer input and calculates its factorial using a for loop.

```python
data = int(input("Enter the number: "))
fact=1
for i in range(1,data+1):
    fact=fact*i
print(f"The factorial of {data} is {fact}")
```

### 3. Reverse Digits of a Number (While Loop) Write a program to reverse the digits of a given integer using a while loop.

```python
data= int(input("Enter the number:"))
while 1<=data:
    print(data, end=" ")
    data-=1
```

### 6. Print All Prime Numbers in a Range (For + While) Write a program to print all prime numbers between 1 and N using a combination of for and while loops.

```python
data = int(input("Enter the value of N: "))

print("Prime numbers between 1 and", data, "are:")

for num in range(2, data + 1):      
    i = 2
    is_prime = True

    while i * i <= num:          
        if num % i == 0:
            is_prime = False
            break
        i += 1

    if is_prime:
        print(num)
```

### Q.7 Fibonacci Series (While Loop) Generate the Fibonacci series up to n terms using a while loop.

```python
data = int(input("Enter the number of terms: "))

# First two terms
a, b = 0, 1
count = 0

print("Fibonacci Series:")

while count < data:
    print(a)
    next_value = a + b
    a = b
    b = next_value
    count += 1
```

### Q.10 Guess the Number Game (While Loop) Create a simple number guessing game where the user keeps guessing until they find the correct number (use while loop).

```python
import random
random_number=random.choice([1,2,3,4,5])
count=1
while True:
    number_guss = eval(input("Enter the number between 1 to 5 = "))
    if number_guss ==random_number:
        print(f"You found it !!!!!!! {number_guss} and {random_number} are MATCHED on attempt of {count}")
        break
    else:
        count+=1
        print("Sorry!!! The input is Wrong or Not Matched")
        continue
```

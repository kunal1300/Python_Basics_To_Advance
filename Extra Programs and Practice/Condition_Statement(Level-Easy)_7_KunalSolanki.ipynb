### Q1: Write a program that takes a number from the user and checks if it is even or odd.



```python
data =eval(input("Please enter any Number: "))
if data%2==0:
    print(f"{data} is Even")
else:
    print(f"{data} is Odd")
```

### Q2: Write a Python program that checks whether a given number is positive, negative, or zero.


```python
data =eval(input("Please enter any Number: "))
if data == 0:
    print(f"{data} is Zero")
elif data>0:
    print(f"{data} is Positive")
else:
    print(f"{data} is Negative")
```

### Q3: Write a program that takes a person's age and prints:
    "Minor" if age < 18
    "Adult" if age is 18 to 59
    "Senior" if age is 60 or more




```python
age =eval(input("Please enter any Age: "))
if age>=60:
    print("Senior")
elif age<=59 and age>18:
    print("Adult")
else:
    print("Minor")
```

### Q4: Write a program that takes two numbers from the user and prints the larger one.



```python
a = float(input("Enter first number: "))
b = float(input("Enter second number: "))

if a > b:
    print("Larger number:", a)
else:
    print("Larger number:", b)
```

### Q5: Write a program that checks whether a given year is a leap year or not.



```python
year = int(input("Enter year: "))

if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print("Leap year")
else:
    print("Not a leap year")
```

### Q7: Write a program to check if a given number is divisible by both 3 and 5.



```python
number = int(input("Enter a number: "))

if number % 3 == 0 and number % 5 == 0:
    print("Divisible by both 3 and 5")
else:
    print("Not divisible by both")
```

### Q8: Write a program that takes marks from the user and prints the grade as per the following:
    • 90 and above: A
    • 80–89: B
    • 70–79: C
    • 60–69: D
    • Below 60: F



```python
marks = float(input("Enter marks: "))

if marks >= 90:
    print("Grade A")
elif marks >= 80:
    print("Grade B")
elif marks >= 70:
    print("Grade C")
elif marks >= 60:
    print("Grade D")
else:
    print("Grade F")
```

### Q9: Write a program to check if a given number is within the range 1 to 100 (inclusive).



```python
num = int(input("Enter a number: "))

if 1 <= num <= 100:
    print("Within range")
else:
    print("Out of range")
```

1. Even or Odd Counter
   Write a program to count how many even and odd numbers are in a given list.



```python
data=[i for i in range(1,11)]
odd=[]
even=[]
count_even=0
count_odd=0

for i in data:
    if i%2==0:
        even.append(i)
        count_even+=1
    else:
        odd.append(i)
        count_odd+=1

print("Original List:",data)
print(f"The number of Even is list is :{count_odd} and number are {even}")
print(f"The number of Odd is list is :{count_odd} and number are {odd}")
```

2. Factorial Using Loop
   Ask the user for a number and calculate its factorial using a for or while loop.



```python
num = eval(input("Enter the number:"))
fact=1
for i in range(1,num+1):
    fact*=i

print(fact)
```

3. Number Range with Conditions
   Print all numbers from 1 to 50, but skip multiples of 5 and stop the loop if the number reaches 35.



```python
data =[x for x in range(1,51)]
print("Original",data)
list1=[]
for i in data:
    if i%5==0:
        if i==35:
            break
    else:
        list1.append(i)
print("Not multiple of 5 and list till 35",list1)
        
```

4. Simple Calculator (if-Elif)
   Write a program that asks for two numbers and an operator (+, -, *, /) and performs the operation using conditional statements.




```python
num1=int(input("Please enter the first number of your choice :"))
num2=int(input("Please enter the second number of your choice :"))

print("The Operation you need to perform \n  1. + \n  2. - \n  3. * \n  4. /  ")

operator=input("Please Enter the operation you need to ferform: ")
if operator == "+":
    print(f"The Addition of {num1} and {num2} is:",num1+num2)
elif operator=="-":
    print(f"The Subtraction of {num1} and {num2} is:",num1-num2)
elif operator=="*":
    print(f"The Multiplication of {num1} and {num2} is:",num1*num2)
elif operator=="/":
    print(f"The Multiplication of {num1} and {num2} is:",num1/num2)
else:
    print(f"The {operator} operator entered is incorrect")
```

5. Check Leap Year
   Take a year as input and check if it is a leap year using conditions.



```python
year = eval(input("Enter the year:"))
if year%4==0:
    if year%400==0 and  year%100!=0:
        print("Not a Leap year")
    else:
        print("Leap year")
else:
    print("Not a Leap year")
```


9. Count Positive, Negative, and Zero Numbers
   From a given list of numbers, count how many are positive, negative, or zero using a loop and conditionals.



```python
numbers = [10,2,3,-3,4,-2,3,0,-4,-45,-3,32,35]

positive_count = 0
negative_count = 0
zero_count = 0

for num in numbers:
    if num > 0:
        positive_count += 1
    elif num < 0:
        negative_count += 1
    else:
        zero_count += 1

print("Positive numbers:", positive_count)
print("Negative numbers:", negative_count)
print("Zeroes:", zero_count)
```

10. Check if a Number is a Palindrome (with strings)
    Check if a number entered by the user is a palindrome using string reversal.



```python
num = input("Enter a number: ")

# Reverse the number using slicing
reversed_num = num[::-1]

if num == reversed_num:
    print(num, "is a palindrome")
else:
    print(num, "is not a palindrome")
```

11. Sum of N Natural Numbers

12. Check Prime (basic)
    Input a number and check if it's a prime number using a for loop and conditionals.


```python
num = int(input("Enter a number: "))

is_prime = True

for i in range(2, num):
    if num % i == 0:
        is_prime = False
        break

print(num,"Prime" if is_prime and num > 1 else "Not Prime")
```

14. Count Vowels in a String
    Input a string and count how many vowels it has using a loop and if statements.

```python
text = input("Enter text: ").casefold()
vowels = "aeiou"
count = 0

for ch in text:
    if ch in vowels:
        count += 1

print(text+" Vowel count:", count)
```

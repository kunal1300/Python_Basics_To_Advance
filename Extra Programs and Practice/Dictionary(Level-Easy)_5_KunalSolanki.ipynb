### 2. **Access a value**  
   Given the dictionary `{'a': 1, 'b': 2, 'c': 3}`, how can you access the value of the key `'b'`?




```python
data={'a': 1, 'b': 2, 'c': 3}
print(data["b"])
```

```python
data.get("b")
```

### 3. **Add a new key-value pair**  
   Add a new key-value pair to the dictionary `{'x': 10, 'y': 20}` where the key is `'z'` and the value is `30`.


```python
data ={'x': 10, 'y': 20}
data["z"]=30
print(data)
```

### 4. **Remove a key-value pair**  
   How would you remove the key-value pair `'b': 2` from the dictionary `{'a': 1, 'b': 2, 'c': 3}`?



```python
data = {'a': 1, 'b': 2, 'c': 3}
data.pop("b")
print(data)
```

### 5. **Check for membership**  
   Check if the dictionary `{'a': 1, 'b': 2, 'c': 3}` contains the key `'b'`.


```python
data={'a': 1, 'b': 2, 'c': 3}
print("b" in data)
```

### 6. **Merge dictionaries**  
   Merge two dictionaries:  
   `dict1 = {'a': 1, 'b': 2}` and `dict2 = {'c': 3, 'd': 4}` into a new dictionary.



```python
dict1 = {'a': 1, 'b': 2}
dict2 = {'c': 3, 'd': 4}
dict3= {**dict1,**dict2}
print(dict3)
```

### 7. **Get all keys**  
   How would you get all the keys from the dictionary `{'name': 'John', 'age': 25}`?



```python
data={'name': 'John', 'age': 25}
print(f"Keys: {list(data.keys())}")
```

### 8. **ASCII dictionary**  
   Create a dictionary with keys as the first letters of the alphabet (e.g., 'a', 'b', 'c') and the corresponding values as their ASCII values.



```python
ascii_dict = {
    'a': ord('a'),
    'b': ord('b'),
    'c': ord('c')
}
print(ascii_dict)
```

### 9. **Update a value**  
   Using the dictionary `{'apple': 5, 'banana': 3}`, update the value of the key `'apple'` to `10`.



```python
data={'apple': 5, 'banana': 3}
data['apple'] = 10
print(data)
```

### 10. **Fetch all values**  
    Given the dictionary `{'name': 'Alice', 'age': 30}`, how can you fetch all the values in the dictionary?

```python
data ={'name': 'Alice', 'age': 30}
print(f"Values: {list(data.values())}")
```


### Q.7 Create a function that checks if a number is positive, negative, or zero.



```python
def check_number(num):
    if num > 0:
        return "Positive"
    elif num < 0:
        return "Negative"
    else:
        return "Zero"
    
print(check_number(5))
print(check_number(-1))
print(check_number(0))
```


### Q.9 Make a function that takes a string and returns how many vowels it has.

```python
def count_vowels(s):
    vowels = "aeiouAEIOU"
    return sum(1 for ch in s if ch in vowels)


data ="Innomatics"
count_vowels(data)
```


### Q.10 Create a function that takes a number and returns True if it's a prime number, otherwise False.

```python
def is_prime(num):
    if num <= 1:
        return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True

is_prime(10)
```

### Q.1 Write a function add(a, b) that returns the sum of two numbers.

```python
def add(a, b):
    return a + b

add(3,5)
```

### Q.2 Write a function is_even(n) that returns True if a number is even, else False.

```python
def is_even(n):
    return n % 2 == 0

is_even(10)
```

### Q.3 Write a function factorial(n) that returns the factorial of a number using recursion.



```python
def factorial(n):
    if n == 0 or n == 1:
        return 1
    return n * factorial(n - 1)

factorial(5)
```

### Q.4 Write a function find_max(a, b, c) that returns the largest among three numbers.



```python
def find_max(a, b, c):
    return max(a, b, c)
  
find_max(10,20,15)
```

### Q.5 Write a function count_vowels(s) that counts the number of vowels in a string.


```python
def count_vowels(s):
    vowels = "aeiouAEIOU"
    return sum(1 for ch in s if ch in vowels)
 
data ="My name is Kunal"
count_vowels(data)
```


### Q.6 Write a function is_palindrome(s) that returns True if the string is a palindrome, else False.


```python
def is_palindrome(s):
    return s == s[::-1]
  

is_palindrome("MOM")
```


### Q.7 Write a function fibonacci(n) that returns a list containing the first n terms of the Fibonacci series.


```python
def fibonacci(n):
    seq = []
    a, b = 0, 1
    for _ in range(n):
        seq.append(a)
        a, b = b, a + b
    return seq
fibonacci(10)
```


### Q.8 Write a function sum_of_digits(n) that returns the sum of all digits in a given number.


```python
def sum_of_digits(n):
    return sum(int(d) for d in str(abs(n)))


sum_of_digits(127)
```


### Q.9 Write a function is_prime(n) that returns True if n is a prime number, else False.


```python
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True
is_prime(37)
```


### Q.10 Write a function reverse_string(s) that returns the reversed version of the input string.

```python
def reverse_string(s):
    return s[::-1]
data ="Innomatics"
reverse_string(data)
```

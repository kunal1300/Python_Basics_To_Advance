**Q1 (Medium):** Write a function that takes a list of strings and returns a new list with all strings converted to uppercase and sorted alphabetically.



```python
def sort_upper_alpha(n):
    print(sorted(list(map(lambda x:str.upper(x),n))))
    
list1=["a","c","e","b","f","d"]
sort_upper_alpha(list1)


```


**Q2 (Medium):** Write a Python program that checks whether a number is a prime or not (input from user).




```python
n=eval(input("Enter the number: "))



flag=True
for i in range(2,n):
    d=2
    while n>d:
        if n%d==0:
            
           flag=False
           break
        d+=1
    if not flag:
        break  
print("Prime" if flag and n > 1 else "Not Prime")

    
```

**Q3 (Medium):** Write a function that takes a list of numbers and returns the second largest number.





```python
def sec_largest_num(n):
    data = sorted(list(map(lambda x:int(x),n)))
    print(data[-2])

list1=[9,5,4,6,1,7,6,3,1,4,8]
sec_largest_num(list1)
```

**Q5 (Medium):** Write a Python function that takes a list of integers and returns a dictionary with the number as the key and its square as the value.




```python
def printdata(n):
    return {x: x**2 for x in n}


list1=[9,8,7,6,5,4,9]
printdata(list1)
```


**Q6 (Medium):** Write a program that takes a list of tuples (each containing name and score), and returns the name of the student with the highest score.





```python
def highscore(n):
    return max(n, key=lambda x:x[1] )


students = [
    ("Kunal", 85),
    ("Amit", 92),
    ("Neha", 78),
    ("Riya", 90)
]
highscore(students)


```

**Q7 (Medium):** Write a function that takes a string and returns a new string with all vowels removed.




```python
def vowels_s(data):
    vowels="AEIOUaeiou"
    list1=[]
    for i in data:
        if i not in vowels:
            list1.append(i)
    print(list1)
        


data ="Innomatics"
vowels_s(data)
```

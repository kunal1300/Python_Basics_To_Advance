### Dictionary (Level-Medium)




---

### 1. Create a dictionary that maps the following keys to values: 'apple' → 3, 'banana' → 5, 'orange' → 2.



```python
fruit={
    "apple" : 3,
    "banana": 5,
    "orange": 2
}
print(fruit)
```

---

### 2. Access the value associated with the key 'banana' from the dictionary:
fruit = {'apple': 3, 'banana': 5, 'orange': 2}

```python
fruit = {'apple': 3, 'banana': 5, 'orange': 2}
print(fruit.get('banana'))
```

---


### 3. Add a new key-value pair 'grape': 4 to the dictionary:
   fruit = {'apple': 3, 'banana': 5, 'orange': 2}



```python
fruit = {'apple': 3, 'banana': 5, 'orange': 2}
fruit["grape"]=4
print(fruit)
```

---

### 4. Update the value of 'orange' to 10 in the dictionary:
   fruit = {'apple': 3, 'banana': 5, 'orange': 2}



```python
fruit = {'apple': 3, 'banana': 5, 'orange': 2}
fruit['orange']=10
print(fruit)
```

---

### 5. Remove the key 'banana' from the dictionary using the del statement:
   fruit = {'apple': 3, 'banana': 5, 'orange': 2}



```python
fruit = {'apple': 3, 'banana': 5, 'orange': 2}
del fruit["banana"]
print(fruit)
```

---

### 6. Check if 'apple' exists as a key in the dictionary:
   fruit = {'apple': 3, 'banana': 5, 'orange': 2}



```python
fruit = {'apple': 3, 'banana': 5, 'orange': 2}
print("apple" in fruit)
```

---

### 7. Get all the keys from the dictionary:
   fruit = {'apple': 3, 'banana': 5, 'orange': 2}



```python
fruit = {'apple': 3, 'banana': 5, 'orange': 2}

print(fruit.keys())
```

---

### 8. Get all the values from the dictionary:
   fruit = {'apple': 3, 'banana': 5, 'orange': 2}


```python
fruit = {'apple': 3, 'banana': 5, 'orange': 2}

print(fruit.values())
```

---


### 9. Clear all items from the dictionary:
fruit = {'apple': 3, 'banana': 5, 'orange': 2}


```python
fruit = {'apple': 3, 'banana': 5, 'orange': 2}

fruit.clear()
print(fruit)
```

### 10. Copy the dictionary into a new variable:
fruit = {'apple': 3, 'banana': 5, 'orange': 2}

```python
fruit = {'apple': 3, 'banana': 5, 'orange': 2}

new_dict = fruit.copy()
new_dict["grape"]=20
print(new_dict)
print(fruit)
```

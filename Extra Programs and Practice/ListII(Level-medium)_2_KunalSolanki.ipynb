### 1. **Create a list** `[10, 20, 30, 40, 50]` and replace the second element with `25`.

```python
list1 =[10,20,30,40,50]
print("Original list:",list1)
list1[1]=25
print("After Replace:",list1)
```

### 2. Given the list `[5, 10, 15, 20, 25]`, **slice** the list to get the elements from index `1` to `3`.

```python
data=[5, 10, 15, 20, 25]
print("Original list:",data)
print("After Slicing:",data[1:4])
```

### 8. **Reverse** the list `[1, 2, 3, 4, 5]` *without* using `reverse()` or slicing `[::-1]`.


```python
data=[1, 2, 3, 4, 5]
rev = []

for i in data:
    rev.insert(0,i)   

print(rev)
```

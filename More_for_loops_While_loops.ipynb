# Given a list of marks find the corresponding result in Pass or Fail, pass if marks>50


```python
# list based iteration
marks_list =[10,20,30,40,70,80,65,98]
grade = []
for i in marks_list:
    if i >= 50:
        grade.append("Pass")
    else:
        grade.append("Fail")
print(grade)
```

```python
# list comprehension makes it concise
["Pass" if i>=50 else "Fail" for i in marks_list]
```

# example of using dictionaries in for loops

```python
dict_marks = {"Alan":100, "Alice":90,"Nidhi":78}
student_grade = {}
for k, v in dict_marks.items():
    grade = "pass" if v >=50 else "fails"
    student_grade[k] = grade
print(student_grade)
```

# same problem with dictionary comprehension

```python
student_grade = {k:"pass" if v>=50 else "fails" for k,v in dict_marks.items()}
student_grade
```

```python
# grade assignment of students based on marks
```

```python
marks = input("Enter the marks of 8 students with space in the range 0 to 100").split(" ")
marks = list(map(float, marks))

grade = []
for mark in marks:
    if mark>90 and mark<=100:
        grade.append("A+")
    elif mark>80 and mark<=90:
        grade.append("A")
    elif mark>70 and mark<=80:
        grade.append("B+")
    elif mark>60 and mark<=70:
        grade.append("B")
    elif mark>50 and mark<=60:
        grade.append("C+")
    elif mark>40 and mark<=50:
        grade.append("C")
    else:
        grade.append("Fail")

marks_grade = dict(zip(marks,grade))
print(marks_grade)
```

# whenever target sum is achieved print the combination


```python
# 40 + 60 =100,  print(40,60) # print the pairs of numbers
target_sum=int(input("Enter the target"))
list_num=[20,40,70,90,10,50,30,80]
# combination in pairs that sum up to target
for i,k in enumerate(list_num): # index, value

    for j in list_num[i+1:]:
        if k+j == target_sum:
            print(k,j)
            break
    break
```

# While

The while statement in Python supports repeated execution of a statement or block of statements that is controlled by a conditional expression.

count = 0 while x > 0: x = x / 2
count += 1 First, expression, which is known as the loop condition, is evaluated. If the condition is false, the while statement ends. If the loop condition is satisfied, the statement or statements that comprise the loop body are executed. When the loop body finishes executing, the loop condition is evaluated again, to see if another iteration should be performed. This process continues until the loop condition is false, at which point the while statement ends.

The loop body should contain code that eventually makes the loop condition false, or the loop will never end unless an exception is raised or the loop body executes a break statement. A loop that is in a functionâ€™s body also ends if a return statement executes in the loop body, as the whole function ends in this case.

The while loop in python is used to run a block of code as long as the condition being checked for is True

Syntax:

while condition: block_of_code i = i + 1 => i += 1

```python
while 5==4+1:
    print("friends forever")
    break
```

```python
# intialize the control variable
c=3
while c>0:
    print("*"*c)
    c-=1
```

```python
x=10
while x>5: # stops when it violates the condition
    print(x)
    x-=1
```

```python
# print odd numbers with while
counter=1
while counter<10:
    if counter%2==1: #  checking a condition
        print(counter)
    counter+=1 # exit when counter=10
        
```

```python
%%writefile vote_processor_1.py
import json
def votes_processor(filepath):
    try:
        with open(filepath,'r') as json_file:
            votes_data=json.load(json_file)
            print(votes_data)
    except FileNotFoundError as fe:
        print(f"error: {fe}")
    except json.JSONDecodeError as je:
        print(f"error: {je}")
    for candidate,votes in votes_data["tally"].items():
        print(f"Candidate {candidate} has got {votes} votes")
    votes_count=votes_data["tally"]
    winner=max(votes_count,key=votes_count.get)
    max_votes=votes_count[winner]
    print(f"The winner is {winner} with {max_votes} votes")
    
```

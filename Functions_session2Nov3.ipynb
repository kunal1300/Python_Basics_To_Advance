# WAP to display if a person is eligible for driving license , take name, location, age and test_passed as parameters

```python
def eligible_license(name,location,age,test_passed):
    ''' eligibility check for driving license''' 
    if age>=18 and test_passed=='Yes':
        print(f"{name} of age {age} based in {location} is eligible for driving license.")
    else:
        print(f"{name} of age {age} based in {location} is not eligible for driving license.")
```

```python
eligible_license("Aditi","Pune",18,"Yes")
```

 *args and **kwargs

*args - when you have no idea about how many arguments you will be receiving as an input for the function
**kwargs - .......how many keyword arguments you will be .......

```python
def average_marks(*marks):
    #number=len(args)
    average=0 #initialize the value we are returning
    print(type(marks))
    if len(marks)>0:
        average=sum(marks)/len(marks)
    else:
        print("There are no students provided")
    return average
```

```python
average_calc=average_marks(10,20,30,75,78)
```

```python
average_calc
```

```python
average_calc=average_marks()
```

```python
average_calc
```

```python
def display_details(**kwargs):
    '''variable length student details passed'''
    print(type(kwargs))
    for k,v in kwargs.items():
        print(f"{k} has value {v}")
        
```

```python
display_details(name="Sakshi",age=18, marks=100,place="Pune")
```

# Recursion

Recursion means calling a function in terms of iteself. All recursive algorithms have these characteristics:

1. A recursive algorithm must call itself, recursively.
2. A recursive algorithm must have a base case. A base case occurs when the problem after repeated calls to the recursive function, becomes so small that it can be solved directly.
3. A recursive algorithm after each call to the recursive function, must move towards the base case. So, in a well-designed recursive algorithm, after each iteration, the problem must become smaller.

```python
def fibonacci(n):
    ''' obtain the nth term of a fibonacci'''
    if n in (0,1):
        return n
    else:
        return fibonacci(n-1)+fibonacci(n-2) # multiple function calls is time taking and memory intensive
```

```python
fibonacci(5)
```

```python
for i in range(7):#7 terms
    print(fibonacci(i),end=" ")
    
```

# print the factorial of a number 

```python
def factorial(n):
    '''factorial of a number'''
    #lowest unit 
    if n==1:
        return n
    else:
        return n*factorial(n-1)
    
```

```python
factorial(5)
```

```python
# sum of the digits of a number
def sum_digit(num):
    if num==0:
        return num
    else:
        last_digit=num%10
        remaining_digit=num//10
        return last_digit + sum_digit(remaining_digit)
```

```python
sum_digit(124)
```

# Some special functions

## 1. zip function
- zip() is an aggregator, meaning it aggregates two or more iterables or sequences. The individual items of the iterables are combined to form tuples. The zip() function stops whenever the shortest of the variables is exhausted.

```python
l1=[1,2,3]
l2=[44,55,66,77,88,99]
zip_list=zip(l1,l2)
print(zip_list)
```

```python
list_zip=list(zip_list)
list_zip
```

```python
#unpack the ziplist
list1,list2= zip(*list_zip)
print(list1)
print(list2)
```

## 2. Lambda function
- Lambda function is used to create an anonymous function (i.e, an inline function without a name) used for a short calculation

### Rules to create Lambda function
- lambda can take any number of parameters (arguments)
- instead of def, we use lambda. No paranthesis required
- colon separates parameters and expression.
- no need of print/return keyword
- parameters/arguments  is optional

```python
# square of numbers
```

```python
lambda x : x*x
```

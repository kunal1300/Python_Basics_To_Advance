# customer feedback analyzer
# You work with a retail company to analyze customer feedback.
# Each feedback has a name, rating (1â€“5), and comment.

```python
feedbacks = [
{"name": "Alice", "rating": 5, "comment": "Great service"},
{"name": "Bob", "rating": 3, "comment": "Averageexperience"},
{"name": "Chetan", "rating": 2, "comment": "Too slow"},
{"name": "Divya", "rating": 4, "comment": "Nice store"},
{"name": "Eshan", "rating": 1, "comment": "Bad service"}
]
```

```python
def classify_feedback(rating):
    """this processes the rating and outputs the sentiment"""
    if rating>=4:
        return "Positive"
    elif rating==3:
        return "Neutral"
    else:
        return "Negative"
```

```python
# output is in the form of a dictionary 
#map transform in the form of sentiment

list_sentiments=list(map(lambda f:classify_feedback(f["rating"]),feedbacks))
```

```python
list_sentiments
```

```python
list_names=list(map(lambda f:f["name"],feedbacks))
list_names
```

```python
sentiments_dict=dict(zip(list_names,list_sentiments))
```

```python
sentiments_dict
```

```python
# filter only the positive sentiments
pos_rev=list(filter(lambda f: f["rating"]>=4,feedbacks))
```

```python
print(len(pos_rev))
```

```python
# calculate the average rating
# show an alert if rating is less than threshold
from functools import reduce
total=reduce(lambda acc,f: acc+f["rating"], feedbacks,0)
print(total)
average_rating=total/len(feedbacks)
print(f"Average rating is {average_rating}")

```

# student data transformation with dictionary comprehension

```python
# 6. Dictionary Comprehension to display the average scores and High performers 
# with average score >=80    
# Student data transformation
students = [
{"name": "Alice", "scores": [85, 90, 88]},
{"name": "Bob", "scores": [70, 75, 72]},
{"name": "Charlie", "scores": [92, 88, 95]}
]
def student_data_transformation():
    dict_avg={k["name"]:round(sum(k["scores"])/len(k["scores"]),2) for k in students}
    #print(dict_avg)
    for k,v in dict_avg.items():
            print(f"Average score of  {k} is : {v}")
    dict_high={k:v for k,v in dict_avg.items() if v>=80}
    print(dict_high)
    

student_data_transformation()
```

# displaying floyd's triangle using nested for while loops 

```python
# user inputs how many rows
num_rows=int(input())
```

```python
num=1
for i in range(1,num_rows+1):
    for j in range(1,i+1):
        print(num, end= " ")
        num+=1
    print()
    
```

```python
num=1
count=1 # rows

while count<=num_rows:
    j=1 # columns
    while j <=count :
        print(num,end=" ")
        num+=1
        j+=1
    count+=1
    print()
    
        
```

# creating a basic ATM withdrawal system

```python
global_balance=2000
```

```python
# atm withdrawal system, pin , amount, balance

def atm_withdrawal(pin,amount):
    """atm withdrawal system"""
    global global_balance

    pin_db = 38900
    if pin_db==pin:
        print("User authenticated")
    else:
        print("Wrong pin, redo the process")
        return
    if amount<=global_balance:
        if amount%100==0:
            global_balance-=amount
            print(f"Amount {amount} is withdrawn")
            print(f"Current balance is {global_balance}")
            return
        
    else:
        print("Amount requested exceeds balance")
        return
        
    
    
    
```

```python
atm_withdrawal(38900,100)
```

# finding out frequency of words in multiple ways 

```python
sentence="I enjoyed the Diwali break, it did not break the momentum however,Diwali was too good. I am good."
```

```python
from collections import Counter
words=sentence.lower().split() # splitiing the sentence into words
word_count=Counter(words)
```

```python
{w:words.count(w) for w in words} #dictionary comprehension
```

```python
word_freq={}
for w in words:
    if w in word_freq:
        word_freq[w]+=1
    else:
        word_freq[w]=1
print(word_freq)
```

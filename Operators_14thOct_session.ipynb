```python
#user inputs a type of operation
#based on that output the result
#basic calculator
first_value=int(input("Enter first value"))
second_value=int(input("Enter second value"))
operator=input("Enter the operation you want to perform:add,sub,mult,div")
if operator == 'add':
    print("Sum:",first_value+second_value)
elif operator == 'sub':
    print("Difference:",first_value-second_value)
elif operator == 'mult':
    print("Product:",first_value*second_value)
else:
    print("Division:",first_value/second_value)
```

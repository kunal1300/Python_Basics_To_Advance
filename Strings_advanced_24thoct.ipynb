```python
str_translate='english'
# Create a translation table using maketrans()
# This maps 'a' to 'x', 'b' to 'y', and 'c' to 'z'
translation_table = str.maketrans("abc", "xyz")

# The original string
original_string = "hello world, abc!"

# Translate the string using the created table
translated_string = original_string.translate(translation_table)

# Print the results
print(f"Original string: {original_string}")
print(f"Translated string: {translated_string}")

# Example with deletion
# Create a translation table to delete 'o' and 'l'
deletion_table = str.maketrans("", "", "ol")

# Translate the string, deleting specified characters
string_with_deletion = "hello world"
deleted_string = string_with_deletion.translate(deletion_table)

print(f"Original string for deletion: {string_with_deletion}")
print(f"String after deletion: {deleted_string}")
```

```python
# String with a printable character
printable_string = "Hello World!"
print(f"'{printable_string}' is printable: {printable_string.isprintable()}")

# String with a newline character (non-printable)
non_printable_string_newline = "Hello\nWorld!"
print(f"'{non_printable_string_newline}' is printable: {non_printable_string_newline.isprintable()}")

# String with a tab character (non-printable)
non_printable_string_tab = "Hello\tWorld!"
print(f"'{non_printable_string_tab}' is printable: {non_printable_string_tab.isprintable()}")

# String with an escape character (ASCII 27)
non_printable_string_escape = chr(27) + "abc"
print(f"'{non_printable_string_escape}' is printable: {non_printable_string_escape.isprintable()}")
```

# reading binary data 

```python
## Example: Reading a file as raw bytes
file_path = 'image1.jpeg'

try:
    with open(file_path, 'rb') as f:
        # Read the entire file content into a bytes object
        raw_data = f.read()
        
    print(f"Successfully read {len(raw_data)} bytes of data.")
    
    # Print the first few bytes (the file signature/header)
    print("\nFirst 16 bytes (Header):")
    print(raw_data[:16])
    
except FileNotFoundError:
    print(f"Error: File not found at {file_path}")
except Exception as e:
    print(f"An error occurred: {e}")
```

```python
from PIL import Image
from io import BytesIO
# Assuming 'raw_data' contains the full 6807 bytes of the file content

# 1. Wrap the binary data in an in-memory file object
image_stream = BytesIO(raw_data)

# 2. Open the image from the stream using Pillow
try:
    img = Image.open(image_stream)
    
    # 3. Display the image
    img.show()
    
    print("Image opened and displayed successfully.")
    
except Exception as e:
    print(f"Error decoding image: {e}")
    print("The data provided may not be a complete or valid JPEG image file.")
```

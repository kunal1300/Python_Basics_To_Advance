# 3. ELSE
- All the cases which aren't satisfied using the if statement are collected in the else block given by the else keyword. else block is the terminal block where all the edge cases are collected. As such, no statement has to be given when we are using the else

```python
cutoffscore = 35
distinctioncutoff = 75 
marksscored = int(input('Enter the Marks:'))

# 3 Scenarios
# - Pass
# - Pass with Distinction
# - Fail


if(marksscored < cutoffscore):
    print('Result: Fail')
elif(marksscored >= distinctioncutoff):
    print('Result: Pass with Distinction')
else:
    print('Result: Pass')
```

## Some example questions

### 1. Write a program which will accept a word from the user and is going to check whether the word starts with a vowel or not

```python
user_input = input()
vowels= 'aeiou'
if user_input.strip().casefold()[0] in vowels: #index is 0 for the starting character
    print("Starts with vowel")
else:
    print("Does not start with vowel")
```

### 2. Write a program which will accept a word and a number from the user. Check if the word starts with a vowel and the number is even and write appropriate conditions accordingly

```python
user_word=input("Enter a word")
user_number=int(input("Enter an integer number"))
if user_word.strip().casefold()[0] in 'aeiou' and user_number%2==0:
    print("Word starts with vowel and number is even") #action is specified in the code block using indentation to identify it as a code block
else:
    print("one or both conditions are not satisfied")
```
